Q27. Convert one decimal number into binary, octal, and hexadecimal in a single program.

```python
decimal = int(input("Enter a decimal number: "))
print(f"Binary: {bin(decimal)}")
print(f"Octal: {oct(decimal)}")
print(f"Hexadecimal: {hex(decimal)}")
```

Q30. Display the binary, octal, and hexadecimal representations of numbers from 1 to 10.

```python
for num in range(1, 11):
    print(
        "Dec:", num,
        "Bin:", bin(num),
        "Oct:", oct(num),
        "Hex:", hex(num)
    )
```
